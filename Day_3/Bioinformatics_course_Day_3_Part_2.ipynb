##**Step 2:** Lets move some .gff files from bactopia-main/annotator/prokka to our Day_3/Part_2/gff_files folder

```python
import shutil
import os
import glob

# Source and destination folders
source_folder = "/content/gdrive/Shareddrives/Bacterial_bioinformatics_workshop/Day_2/Part_1/Bactopia_results"
destination_folder = "/content/gdrive/Shareddrives/Bacterial_bioinformatics_workshop/Day_3/Part_2/gff_files"

# Search for all .gff.gz files in the source folder and its subdirectories
files_to_copy = [f for f in glob.glob(os.path.join(source_folder, "**/*.gff.gz"), recursive=True) if os.path.isfile(f)]

# Copy each file to the destination folder directly
for source_path in files_to_copy:
    # Construct the destination path by combining the destination folder and the filename
    destination_path = os.path.join(destination_folder, os.path.basename(source_path))

    # Copy the file
    shutil.copy(source_path, destination_path)
    print(f"Copied {source_path} to {destination_path}")
```

```python
import gzip
import shutil
import os

# Folder containing the .gff.gz file
folder_path = "/content/gdrive/Shareddrives/Bacterial_bioinformatics_workshop/Day_3/Part_2/gff_files"

# List of .gff.gz files to extract
files_to_extract = [
    "SRX2195078.gff.gz",
    "SRX2195079.gff.gz",
    "SRX2195080.gff.gz",
    "SRX2195081.gff.gz",
    "SRX2195082.gff.gz",
    "SRX2195083.gff.gz",
    "SRX2195085.gff.gz",
    "SRX2195086.gff.gz",
    "SRX2195087.gff.gz",
    "SRX2195088.gff.gz",
    # Add more files as needed
]

# Extract each file
for file_name in files_to_extract:
    file_path = os.path.join(folder_path, file_name)
    output_path = os.path.join(folder_path, file_name.replace('.gz', ''))

    # Check if the file exists in the folder before extracting
    if os.path.exists(file_path):
        with gzip.open(file_path, 'rb') as f_in:
            with open(output_path, 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)
        print(f"Extracted {file_name} in {folder_path}")
    else:
        print(f"{file_name} not found in {folder_path}")
```
